```python
from google.colab import drive
drive.mount('/content/drive') #連雲端硬碟

import pandas as pd
googlepath='/content/drive/MyDrive/巨資雲端技術/' #路徑到指定資料夾
```

```python
df = pd.read_csv(googlepath + "初始資料.csv") #讀取檔案
df = df.drop("週期",axis=1) #刪除週期欄位
df = df.drop("波向",axis=1) #刪除波向欄位
df = df.drop("風向",axis=1) #刪除風向欄位
df #顯示資料框內容
```

```python
import numpy as np
'''在資料分析（特別是氣象、水文、醫療或舊型資料庫）中，
常常會用特定的數值（如-9.9、-999或-99）來代表「缺失值」或「異常值」。'''
df.replace(-9.9, np.nan, inplace=True) #將所有的 -9.9 替換為標準的 NaN
df = df.dropna() #刪除有缺失值的整列
df
```

```python
df_1 = df[df['月'] == 1] #只保留「月」欄位等於 1 的整列資料
df_1.to_csv(googlepath + "1月.csv", index=False, encoding="utf-8-sig") #結合路徑並輸出CSV檔
df_1
```
